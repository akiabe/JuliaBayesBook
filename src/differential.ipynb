## 1変数関数の微分

```julia
# 2次関数を定義
f(x) = -(x + 1) * (x - 1)
```

```julia
# 微小な変化量を設定
h = 1.0e-10;
```

```julia
# 導関数fʼの近似式
fʼ(a) = (f(a + h) - f(a)) / h # f\rasp<TAB>
```

```julia
# 可視化範囲の設定
xs = range(-1, 1, length=100);
```

```julia
# 関数、導関数の可視化
plot(
    plot(xs, f.(xs), xlab="x", ylab="y", title="function f(x)", label=:none),
    plot(xs, fʼ.(xs), xlab="x", ylab="y", title="function fʼ(x)", label=:none)
)
```

## 自動微分

```julia
using ForwardDiff
```

```julia
# 2次関数を定義
f(x) = -(x + 1) * (x - 1)
```

```julia
# 導関数を求める
fʼ(x) = ForwardDiff.derivative(f, x)
```

```julia
# 可視化範囲を設定
xs = range(-1, 1, length=100);
```

```julia
plot(
    plot(xs, f.(xs), xlab="x", ylab="y", title="function f(x)", label=:none),
    plot(xs, fʼ.(xs), xlab="x", ylab="y", title="function fʼ(x)", label=:none)
)
```

$f(x) = sin(x)$の導関数を計算する。

```julia
xs = range(0, 2π * 3, length=100)

f(x) = sin(x)
fʼ(x) = ForwardDiff.derivative(f, x)

plot(xs, f.(xs), xlab="x", ylab="y", label="function f(x)")
plot!(xs, fʼ.(xs), xlab="x", ylab="y", label="function fʼ(x)")
```

$sig(x) = \frac{1}{1 + exp(-x)}$の導関数を計算する。

```julia
xs = range(-5, 5, length=100)

f(x) = 1 / (1 + exp(-x))
fʼ(x) = ForwardDiff.derivative(f, x)

plot(xs, f.(xs), xlab="x", ylab="y", label="function f(x)")
plot!(xs, fʼ.(xs), xlab="x", ylab="y", label="function fʼ(x)")
```
