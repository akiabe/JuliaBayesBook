# ベクトル場の可視化

参考：https://qiita.com/Lirimy/items/4cb7adf149d75753bf25

```julia
using Plots
```

```julia
# 可視化する範囲を設定
xs = range(-1, 1, length=15)
ys = range(-1, 1, length=15);
```

```julia
# 1次元の座標系を設定
x = repeat(xs, outer=length(ys))
y = repeat(ys, inner=length(xs));
```

```julia
# ベクトル場の作成
u = y
v = x;
```

```julia
# ベクトル場の可視化
k = 0.1
quiver(x, y, quiver=(k * u, k * v))
```

### Base.repeatの使い方

```julia
repeat([1, 2, 3], 2)
```

```julia
repeat([1, 2, 3], 2, 3)
```

```julia
repeat(1:2, inner=2)
```

```julia
repeat(1:2, outer=2)
```

```julia
# 縦にinnerで2回、横にouterで3回repeatする
repeat([1 2; 3 4], inner=(2, 1), outer=(1, 3))
```
